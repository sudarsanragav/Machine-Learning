### Problem Statement:                                                                                                                     
 #### The car company that produces new model SUV car.                                                                        
 #### The task is to predict that already purchased SUV customers will buy this new SUV or not.                                
 #### Then the prediction report will be send to the Social Marketing team to publish ads on Social Network about New Car.     

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Dataset

```python
data = pd.read_csv('Social_Network_Ads.csv')
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

### Checking for outlier


```python
fig, g = plt.subplots(1,2, figsize = (12,4))
g1 = sns.boxplot(data = data,y= 'Age', color = 'coral',ax = g[0])
g2 = sns.boxplot(data = data,y= 'EstimatedSalary', color = 'coral',ax = g[1])
```

There is no outlier in the Dataset

### Splitting the Dataset into Dependent and Independent 

```python
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
x
```

### Splitting the Data into Train and Test set


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=0)
```

```python
print(x_train)
```

```python
x_train.shape
```

```python
y_train.shape
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
#here we are doing the feature scaling in only independent Variables 'Age' and "EstimatedSalary"
```

```python
print(x_train)
```

```python
print(x_test)
```

### Training the Logistic Regression Model on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

### Predicting New result

```python
print(classifier.predict(sc.transform([[30,87000]]))) #sinle Observation Prediction
# here we are using scaling method to find the prediction 'sc.transform()'
```

### Predicting the results with Test Set

```python
y_prediction = classifier.predict(x_test)
print(np.concatenate((y_prediction.reshape(len(y_prediction),1),y_test.reshape(len(y_test),1)),1))
```

### Making the Confusion Matrix and Accuracy Score

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_prediction)
ac_score = accuracy_score(y_test, y_prediction)
# accuracy score is the rate of correct Prediction
```

```python
print(f'Confusion Matrix: {cm}\nAccuracy Score: {ac_score}')
```
